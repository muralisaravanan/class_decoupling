```python
fluid_k0 = np.loadtxt("fluid_Nddec1perturbations_k0_s.dat")
fluid_k1 = np.loadtxt("fluid_Nddec1perturbations_k1_s.dat")
fluid_k2 = np.loadtxt("fluid_Nddec1perturbations_k2_s.dat")
fluid_k3 = np.loadtxt("fluid_Nddec1perturbations_k3_s.dat")

fluid_k0_new_ICS = np.loadtxt("fluid_Nddec1_new_ICSperturbations_k0_s.dat")
fluid_k1_new_ICS = np.loadtxt("fluid_Nddec1_new_ICSperturbations_k1_s.dat")
fluid_k2_new_ICS = np.loadtxt("fluid_Nddec1_new_ICSperturbations_k2_s.dat")
fluid_k3_new_ICS = np.loadtxt("fluid_Nddec1_new_ICSperturbations_k3_s.dat")

free_k0 = np.loadtxt("freestreaming_Nddec1perturbations_k0_s.dat")
free_k1 = np.loadtxt("freestreaming_Nddec1perturbations_k1_s.dat")
free_k2 = np.loadtxt("freestreaming_Nddec1perturbations_k2_s.dat")
free_k3 = np.loadtxt("freestreaming_Nddec1perturbations_k3_s.dat")
```

```python
fluid_k0_tau = abs(fluid_k0[:,0])
fluid_k1_tau = abs(fluid_k1[:,0])
fluid_k2_tau = abs(fluid_k2[:,0])
fluid_k3_tau = abs(fluid_k3[:,0])

fluid_k0_new_ICS_tau = abs(fluid_k0_new_ICS[:,0])
fluid_k1_new_ICS_tau = abs(fluid_k1_new_ICS[:,0])
fluid_k2_new_ICS_tau = abs(fluid_k2_new_ICS[:,0])
fluid_k3_new_ICS_tau = abs(fluid_k3_new_ICS[:,0])

free_k0_tau = abs(free_k0[:,0])
free_k1_tau = abs(free_k1[:,0])
free_k2_tau = abs(free_k2[:,0])
free_k3_tau = abs(free_k3[:,0])

fluid_k0_delta_g = abs(fluid_k0[:,2])
fluid_k1_delta_g = abs(fluid_k1[:,2])
fluid_k2_delta_g = abs(fluid_k2[:,2])
fluid_k3_delta_g = abs(fluid_k3[:,2])

fluid_k0_new_ICS_delta_g = abs(fluid_k0_new_ICS[:,2])
fluid_k1_new_ICS_delta_g = abs(fluid_k0_new_ICS[:,2])
fluid_k2_new_ICS_delta_g = abs(fluid_k0_new_ICS[:,2])
fluid_k3_new_ICS_delta_g = abs(fluid_k0_new_ICS[:,2])

free_k0_delta_g = abs(free_k0[:,2])
free_k1_delta_g = abs(free_k1[:,2])
free_k2_delta_g = abs(free_k2[:,2])
free_k3_delta_g = abs(free_k3[:,2])

fluid_k0_delta_cdm = abs(fluid_k0[:,12])
fluid_k1_delta_cdm = abs(fluid_k1[:,12])
fluid_k2_delta_cdm = abs(fluid_k2[:,12])
fluid_k3_delta_cdm = abs(fluid_k3[:,12])

free_k0_delta_cdm = abs(free_k0[:,15])
free_k1_delta_cdm = abs(free_k1[:,15])
free_k2_delta_cdm = abs(free_k2[:,15])
free_k3_delta_cdm = abs(free_k3[:,15])

fluid_k0_delta_ddec = abs(fluid_k0[:,18])
fluid_k1_delta_ddec = abs(fluid_k1[:,18])
fluid_k2_delta_ddec = abs(fluid_k2[:,18])
fluid_k3_delta_ddec = abs(fluid_k3[:,18])

fluid_k0_new_ICS_delta_ddec = abs(fluid_k0_new_ICS[:,18])
fluid_k1_new_ICS_delta_ddec = abs(fluid_k0_new_ICS[:,18])
fluid_k2_new_ICS_delta_ddec = abs(fluid_k0_new_ICS[:,18])
fluid_k3_new_ICS_delta_ddec = abs(fluid_k0_new_ICS[:,18])

free_k0_delta_ddec = abs(free_k0[:,17])
free_k1_delta_ddec = abs(free_k1[:,17])
free_k2_delta_ddec = abs(free_k2[:,17])
free_k3_delta_ddec = abs(free_k3[:,17])
```
